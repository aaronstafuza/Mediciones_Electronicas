```
import numpy as np
#medicion de resistencia de 100k cbm
v = 9.044
i=89*10**-6
n=5

v_std=0.015
i_std=5e-6

ra=100
r=v/i
print("R medida:",r)
Ev=0.05+(1/9044)*100
Ea=0.8+(1/89)*100

#incertidumbre tipo A
ui_v=v_std/np.sqrt(n-1)
ui_i=i_std/np.sqrt(n-1)

#incertidumbre asociada a la medicion
dR_dV=1/i
dR_dI=-v/(i**2)

error_V=Ev*v/100
error_I=Ea*i/100

vj=error_V/np.sqrt(3)
ij=error_I/np.sqrt(3)

uc_v=np.sqrt(ui_v**2+vj**2)
uc_i=np.sqrt(ui_i**2+ij**2)

#ratio
ratio_v=vj/ui_v
print("Ratio V:",ratio_v)
ratio_i=ij/ui_i
print("Ratio I:",ratio_i)

uR=np.sqrt((uc_v*dR_dV)**2+(uc_i*dR_dI)**2)
k=2
print("R medida: %f +- %f"%(r,k*uR))

Em_cbm=ra/(r-ra)
print(Em_cbm)
if Em_cbm>((uR/r)*k*0.1):
  print("Se debe corregir")
else:
  print("No hace falta corregir")

```
