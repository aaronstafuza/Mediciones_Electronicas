```
import numpy as np
e=1 #error del instrumental porcentual
v=np.array([5.007,4.994,5.005,4.990,4.999])
i=np.array([19.663,19.639,19.64,19.685,19.678])/1000

#medias
v_med=np.mean(v) 
print("Media tension: ",v_med)
i_med=np.mean(i)
print("Media corriente: ",i_med)

#desvios
v_std=np.std(v)
print("Desviacion de tension: ",v_std)
i_std=np.std(i)
print("Desviacion de corriente: ",i_std)

#incertidumbres
v_i=v_std/np.sqrt(len(v)-1)
print("Incertidumbre (tipo A) de tension: ",v_i)
i_i=i_std/np.sqrt(len(i)-1)
print("Incertidumbre (tipo A) de corriente: ",i_i)

#mientra mas sigma tomamos es mas pesimista pero me aseguro que el valor este dentro 
#de esa ventana

#Errores
error_v=(e/100)*(v_med)
print("Error tension: ",error_v)
error_i=(e/100)*(i_med)
print("Error corriente: ",error_i)

#Paso los errores a incertidumbre
v_j=error_v/np.sqrt(3)
print("Incertidumbre (tipo B) tension: ",v_j)
i_j=error_i/np.sqrt(3)
print("Incertidumbre (tipo B) corriente: ",i_j)

#Incertidumbre combinada
uc_v=np.sqrt(v_i**2+v_j**2)
print("Incertidumbre combinada V: ",uc_v)
uc_i=np.sqrt(i_i**2+i_j**2)
print("Incertidumbre combinada I: ",uc_i)

#Grado efectivo de libertad
veff_v=(uc_v**4)/(v_i**4)
print("Grado efectivo de libertad de tension: ",veff_v)
veff_i=(uc_i**4)/(i_i**4)
print("Grado efectivo de libertad de corriente: ",veff_i)

ratio_v=v_i/v_j
print("Ratio V: ",ratio_v) 
ratio_i=i_i/i_j
print("Ratio I: ",ratio_i)

#por tabla M3003, Kv y Ki es 1.66

Kv=1.66
Ki=1.66

#para que sirven los K? 
#el factor de cobertura k para una probabilidad de cobertura de 95,45
#NUNCA expandir en el medio del problema

print("Tension medida: %f +- %f con factor de cobertura del 95%%"%(v_med,uc_v*Kv))
print("Corriente medida: %f +- %f con factor de cobertura del 95%%"%(i_med,uc_i*Ki))
print("\n")

#Calculo resistencia: r=v/i
R_med=v_med/i_med
print("Resistencia media: ",R_med)
dR_dI=v_med/(i_med**2)
dR_dV=1/i_med
uR=np.sqrt(((dR_dI*uc_i)**2)+(dR_dV*uc_v)**2)
print("Incertidumbre sin expandir R: ",uR)
#Por teorema central del limite vamos a decidir expandir por:

```
