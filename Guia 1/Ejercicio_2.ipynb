```
import numpy as np
e=2 #error del instrumental porcentual
v=np.array([5.007,4.994,5.005,4.990,4.999])
i=np.array([19.663,19.639,19.640,19.685,19.678])/1000
tita=np.array([1.0456,1.0438,1.0468,1.0428,1.0433])

#medias (dato)
v_media=4.9990
i_media=19.6610/1000
tita_media=1.04446

#desvios (dato)
si_v=0.0032
si_i=0.0000095
si_tita=0.00075

#los coeficientes de correlacion del ejercicio son: (dato)
c_vi=-0.36
c_vtita=0.86
c_itita=-0.65

#incertidumbres tipo A
vi=si_v/np.sqrt(len(v)-1)
ii=si_i/np.sqrt(len(i)-1)
titai=si_tita/np.sqrt(len(tita)-1)

#errores
error_v=(e*v_media)/100
error_i=(e*i_media)/100
error_tita=(e*tita_media)/100

#paso errores a incertidumbre 
vj=error_v/np.sqrt(3) 
ij=error_i/np.sqrt(3)
titaj=error_tita/np.sqrt(3)

#calculo incertidumbre combinada
uc_v=np.sqrt(vi**2+vj**2)
uc_i=np.sqrt(ii**2+ij**2)
uc_tita=np.sqrt(titai**2+titaj**2)

#calculo media de R,X,Z
R=(v_media/i_media)*np.cos(tita_media)
X=(v_media/i_media)*np.sin(tita_media)
Z=(v_media/i_media)*np.cos(tita_media)+ 1j*(v_media/i_media)*np.sin(tita_media)

print("\nDATOS PEDIDOS EN EL ENUNCIADO:")
#calculo la incertidumbre comibada de la resistencia
dR_dV=(1/i_media)*np.cos(tita_media)
dR_dI=(-v_media/(i_media**2))*np.cos(tita_media)
dR_dTita=-(v_media/i_media)*np.sin(tita_media)
uR=np.sqrt((uc_v*dR_dV)**2 + (uc_tita*dR_dTita)**2 + (uc_i*dR_dI)**2 + 2*c_vi*dR_dV*uc_v*dR_dI*uc_i + 2*c_vtita*dR_dV*uc_v*dR_dTita*uc_tita + 2*c_itita*dR_dI*uc_i*dR_dTita*uc_tita)
k=2 #por teorema central del limite
print("Resistencia con factor de cobertura del 95%% = %.2f +- %.2f"%(R,uR*k))

#calculo la incertidumbre combinada de la reactancia
dX_dV=(1/i_media)*np.sin(tita_media)
dX_dI=(-v_media/i_media**2)*np.sin(tita_media)
dX_dTita=(v_media/i_media)*np.cos(tita_media)
uX=np.sqrt(((uc_v*dX_dV)**2) + ((uc_i*dX_dI)**2) + ((uc_tita*dX_dTita)**2) + 2*c_vi*dX_dV*uc_v*dX_dI*uc_i + 2*c_vtita*dX_dV*uc_v*dX_dTita*uc_tita + 2*c_itita*dX_dI*uc_i*dX_dTita*uc_tita)
print("Reactancia con factor de cobertura del 95%% = %.2f +- %.2f"%(X,uX*k))

#calculo la incertidumbre combinada de la impedancia
dZ_dV=(1/i_media)*np.cos(tita_media)+ 1j*(1/i_media)*np.sin(tita_media)
dZ_dI=(-v_media/(i_media**2))*np.cos(tita_media)+1j*(-v_media/i_media**2)*np.sin(tita_media)
dZ_dTita=-(v_media/i_media)*np.sin(tita_media)+1j*(v_media/i_media)*np.cos(tita_media)
uZ=np.sqrt(((uc_v*dZ_dV)**2) + ((uc_i*dZ_dI)**2) + ((uc_tita*dZ_dTita)**2) + 2*c_vi*dZ_dV*uc_v*dZ_dI*uc_i + 2*c_vtita*dZ_dV*uc_v*dZ_dTita*uc_tita + 2*c_itita*dZ_dI*uc_i*dZ_dTita*uc_tita)
#tomo k=2 por teorema central del limite
print("Impedancia con factor de cobertura del 95%%= ",Z,end="+- ")
print(uZ*k)



```
