```
import numpy as np
v1 = 100.145 #media 
si = 1.489 #desviacion estandar
error_porcentual = 0.5+3/1001*100 #error del instrumento

#calculo incertidumbre A
ui=si/np.sqrt(19) #19: cantidad total de mediciones - 1 
print("Incertidumbre (tipo A): ",ui)

#calculo incertidumbre B
error = v1*error_porcentual/100
uj=error/np.sqrt(3)
print("Incertidumbre (tipo B): ",uj)

#calculo incertidumbre combinada
uc=np.sqrt(ui**2+uj**2)
print("Incertidumbre combinada: ",uc)

#grados de libertad
v_eff=(uc**4)/(ui**4)
print("Grados efectivos de libertad: ", v_eff)

#calculo ratio
r = ui/uj
print("Ratio: ",r)
#para ver que factor de cobertura obtengo, me fijo en la tabla m3003. 
k=1.91
print("Tension que obtuve con factor de cobertura del 95%% = %f ± %f: "%(v1,uc*k))
```
