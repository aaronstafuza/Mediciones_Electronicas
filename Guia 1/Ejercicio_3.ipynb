```
import numpy as np
i_amperimetro = np.array([10.01,10.00,10.02,10.01,10.00]) #mediciones de corrientes tomadas con el amperimetro
i_calibrador = 10 
i_media = np.mean(i_amperimetro) #corriente media del amperimetro
error_porcentual = 1 #del amperimetro

#desvio del amperimetro
si_amperimetro = np.std(i_amperimetro) 
print("Desvio del amperimetro: ",si_amperimetro)

#incertidumbre (tipo A) del amperimetro
ui_amperimetro = si_amperimetro/np.sqrt(len(i_amperimetro)-1) 
print("Incertidumbre (tipo A) amperimetro: ", ui_amperimetro)

#incertidumbre (tipo B) del amperimetro
error_amperimetro = (i_media*error_porcentual)/100
print("Error amperimetro: ",error_amperimetro)
uj_amperimetro = error_amperimetro/np.sqrt(3)
print("Incertidumbre (tipo B) amperimetro: ",uj_amperimetro)

#incertidumbre combinada del amperimetro
uc_amperimetro = np.sqrt(ui_amperimetro**2 + uj_amperimetro**2)
print("Incertidumbre combinada del amperimetro: ",uc_amperimetro)

error = i_media - i_calibrador
de_dimedia=1
de_dicalibrador=-1

#voy a necesitar la incertidumbre B del calibrador, la calculo:
error_calibrador = (i_calibrador*0.05/100)+0.002
uj_calibrador = error_calibrador/3
print("Incertidumbre (tipo B) calibrador: ",uj_calibrador)

#calculo incertidumbre del error
u_error = np.sqrt((de_dimedia*uc_amperimetro)**2+(de_dicalibrador*uj_calibrador)**2)
print("Incertidumbre del error: ",u_error)

k=2 #por teorema central del limite
print("Error efectuado en la medida con k del 95%%: %f ± %f"%(error,u_error*k))


```
