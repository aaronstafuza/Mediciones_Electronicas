```
import numpy as np
Rp=10e6
Vrx=10.134
error_Vrx=0.05+(2/10134)*100
Vrp=5.637
error_Vrp=0.05+(2/5637)*100

#A
ui_Rp=0.01e6/2
ui_Vrx=4.8e-3
ui_Vrp=3.9e-3

#B
eVrp=(error_Vrx*Vrx)/100
eVrx=(error_Vrp*Vrp)/100

uj_Vrx=eVrx/np.sqrt(3)
uj_Vrp=eVrp/np.sqrt(3)
#combinada
uc_Vrx=np.sqrt(ui_Vrx**2+uj_Vrx**2)
uc_Vrp=np.sqrt(ui_Vrp**2+uj_Vrp**2)

Rx=Rp*(Vrx/Vrp)
dRx_dRp=Vrx/Vrp
dRx_dVrx=Rp/Vrp
dRx_dVrp=-Rp*Vrx/(Vrp**2)
uRx=np.sqrt((dRx_dRp*ui_Rp)**2+(dRx_dVrx*uc_Vrx)**2+(dRx_dVrp*uc_Vrp)**2)
print("R medida = %f +- %f"%(Rx,uRx*2))

print(100*ui_Rp/Rp)
print(100*uc_Vrx/Vrx)
print(100*uc_Vrp/Vrp)
print(100*ui_Rp/Rp+100*uc_Vrx/Vrx+100*uc_Vrp/Vrp)
```
