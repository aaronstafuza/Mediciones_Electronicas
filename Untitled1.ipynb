```
import numpy as np

V=1.35
I=979e-6
T=10*200e-6
t_desfase=1.25*200e-6

#incertidumbres:
uc_v=V*0.004
uc_i=I*0.004
error_T=((200e-6)/8000+T*(T*50/1e6)+0.6e-9)/np.sqrt(3)
error_tdesfase=((200e-6)/8000+t_desfase*(t_desfase*50/1e6)+0.6e-9)/np.sqrt(3)

#resistencia
R=(V/I)*np.cos(t_desfase*2*np.pi/T)
#incertidumbre, calculo las parciales
dR_dV=(1/I)*np.cos(t_desfase*2*np.pi/T)
dR_dI=-(V/(I**2))*np.cos(t_desfase*2*np.pi/T)
dR_dT=-(V/I)*np.sin(t_desfase*2*np.pi/T)*(-2*np.pi/(T**2))
dR_dtdes=-(V/I)*np.sin(t_desfase*2*np.pi/T)*2*np.pi/T
uR=np.sqrt((uc_v*dR_dV)**2+(uc_i*dR_dI)**2+(error_T*dR_dT)**2+(error_tdesfase*dR_dtdes)**2)
k=2
print("R = %f +- %f"%(R,uR*k))

#reactancia
Xc=(V/I)*np.sin(t_desfase*2*np.pi/T)
#incertidumbre, calculo las parciales
dX_dV=(1/I)*np.sin(t_desfase*2*np.pi/T)
dX_dI=-(V/(I**2))*np.sin(t_desfase*2*np.pi/T)
dX_dT=(V/I)*np.cos(t_desfase*2*np.pi/T)*(-2*np.pi/(T**2))
dX_dtdes=(V/I)*np.cos(t_desfase*2*np.pi/T)*2*np.pi/T
uX=np.sqrt((uc_v*dX_dV)**2+(uc_i*dX_dI)**2+(error_tdesfase*dX_dtdes)**2+(error_T*dX_dT)**2)
print("Xc = %f +- %f"%(Xc,k*uX))

#capacitancia
C=T/(2*np.pi*Xc)
dC_dT=1/(2*np.pi*Xc)
dC_dX=-T/(2*np.pi*Xc**2)
uC=np.sqrt((error_T*dC_dT)**2+(uX**dC_dX)**2)
print("C = %f +- %f"%(C*1e9,k*uC*1e9))
```
