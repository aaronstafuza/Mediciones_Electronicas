```
import numpy as np

Vc=0.953  #v media en continua
Va=1.057 #v media en alterna


Rth=500e3
Rv=1e6

si_vC=0.10e-3 #desvio en continua
si_vA=0.25e-3 #desvio en alterna

eC= 0.8+1/953*100 #error DC
eA= 1+5/1057*100 #error AC

#incertidumbre tipo A
ui_vc=si_vC/np.sqrt(4) 
ui_va=si_vA/np.sqrt(4)

#errores
error_vC=(eC*Vc)/100
error_vA=(eA*Va)/100

#errores a incertidumbre
uj_vc=error_vC/np.sqrt(3)
uj_va=error_vA/np.sqrt(3)

#incertidumbre combinada
uc_vc=np.sqrt(ui_vc**2+uj_vc**2)
uc_va=np.sqrt(ui_va**2+uj_va**2)

#error de metodo de Vc
Em=(1/(1+(Rth/Rv)))-1
print("Error de metodo de Vc: ",Em)

#calculo Vc
Vc_corregida=(Vc/(1+Em))
print("Vc corregida: %.3fV"%Vc_corregida)
print("Incertidumbre de Vc = %.3f"%uc_vc)
print("\n")

#calculo Va
Va_corregida = (Va/(1+Em))
print("Va corregida: %.3fV"%Va_corregida)
print("Incertidumbre de Va: %.3f"%uc_va)
print("\n")

Vme=Va_corregida/(np.pi/(2*np.sqrt(2)))
uj_Vme=error_vA/np.sqrt(3) #incertidumbre tipo B Vme

Vp_pos =2*Vme+Vc_corregida #formula Vp+
Vp_neg=Vc_corregida-2*Vme #formula Vp-

#calculo la incertidumbre de Vp+
dVpos_dVme=2
dVpos_dVc=1
uV_pos=((uc_vc*dVpos_dVc)**2+(uj_Vme*dVpos_dVme)**2)
k=2 #por teorema central del limite
print("Vp+ con confianza de 95%%: %.3fV +- %.4fV"%(Vp_pos,uV_pos*k))

#calculo la incertidumbre de Vp-
dVpneg_dVme=-2
dVpneg_dVc=1
uV_neg=((uc_vc*dVpneg_dVc)**2+(uj_Vme*dVpneg_dVme)**2)
print("Vp- con confianza de 95%%: %.3fV +- %.4fV"%(Vp_neg,uV_neg*k))


```
