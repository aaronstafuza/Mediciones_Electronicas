```
import numpy as np
Rg=600
Rv=1e6
Cv=200e-12
Cd=2e-12
l=90
Cc=Cd*l
f=5000
w=2*np.pi*f

#las ecuaciones utilizadas en este codigo estan en el pdf Guia 3 
#calculo error de metodo
Em=(Rv/(Rv+Rg)*(1/np.sqrt(1+(w*(Cc+Cv)*Rg*Rv/(Rv+Rg))**2))-1)
print("Error de metodo: %.5f"%Em)

Vi=1.9805 #tension medida
error_v=0.05+(2/19805)*100

e_v=Vi*error_v/100
uj_v=e_v/np.sqrt(3)

k=2 #coef para cobertura 95%
print("Error porcentual de la incertidumbre = %.4f%%"%(uj_v*k*100/Vi))
print("Uexp(Vv): %.4f"%(uj_v*k))

if abs(Em)>((uj_v*k/Vi)/10):
  print("Se debe corregir")
else:
  print("No hace falta corregir")

Vg=Vi/(Em+1)
print("Tension corregida: %.4fV +- %.4fV"%(Vg,uj_v*k))

```
