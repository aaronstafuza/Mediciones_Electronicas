```
import numpy as np
i=np.array([145.1,145.2,144.9,145.7,145.2,146.3,145.3,145.1,145.9,145.4])/1000
v=np.array([22.323,22.325,22.320,22.331,22.332,22.330,22.327,22.323,22.329,22.325])

v_media=np.mean(v)
print("Tension media: ",v_media)
i_media=np.mean(i)
print("Corriente media: ",i_media*1000) #en mA

r=v_media/i_media

#error voltimetro y su resistencia:
error_volt=0.05+(2/223265)*100
Rv=10e6

#error amperimetro y su resistencia:
error_amp=0.15*0.2/100
Ra=100e-3

#coeficiente de correlacion calculado en Excel:
coef_correl=0.67

#desvios de corriente y tension:
v_std=np.std(v)
i_std=np.std(i)

#incertidumbre tipo A:
ui_v=v_std/np.sqrt(len(v)-1)
ui_i=i_std/np.sqrt(len(i)-1)

#errores:
errorV=(error_volt*v_media)/100
errorI=(error_amp*i_media)/100

#errores a incertidumbre tipo B:
uj_v=errorV/np.sqrt(3)
uj_i=errorI/np.sqrt(3)

#incertidumbre combinada:
uc_v=np.sqrt(ui_v**2+uj_v**2)
print("Incertidumbre de la tension: %.4f"%uc_v)
uc_i=np.sqrt(ui_i**2+uj_i**2)
print("Incertidumbre de la corriente: %.5f"%uc_i)

#incertidumbre asociada a la medicion:
dR_dV=1/i_media
dR_dI=-v_media/(i_media**2)
uR=np.sqrt((uc_v*dR_dV)**2+(uc_i*dR_dI)**2+ 2*coef_correl*dR_dV*uc_v*dR_dI*uc_i)
k=2 #por teorema central del limite
print("R medida: %f ± %f"%(r,k*uR))

Em_tbm=-r/Rv
if abs(Em_tbm)>((uR/r)*k*0.1):
  print("Se debe corregir")
else:
  print("No hace falta corregir")

#En caso de necesitar corregir la ecuacion utilizada seria:
# Vg=v_media/(Em_tbm+1)
#print("Tension corregida: %f +- %f"%(Vg,uc_v*k))
```
