```
import numpy as np

V1=0.55*2
V2=0.85*2
P1=40
e_P1=5

error_P1=P1*e_P1/100
u_P1=error_P1/np.sqrt(3)
u_V1=V1*0.02/np.sqrt(3)
u_V2=V2*0.02/np.sqrt(3)

#potencia de envolvente
Px=P1*(V2/V1)**2
dPx_dP1=(V2/V1)**2
dPx_dV1=-2*P1*(V2**2)/(V1**3)
dPx_dV2=2*P1*V2/(V1**2)
uPx=np.sqrt((u_P1*dPx_dP1)**2+(u_V1*dPx_dV1)**2+(u_V2*dPx_dV2)**2)
k=2
print("Px = %f+-%f"%(Px,uPx*k))
```
