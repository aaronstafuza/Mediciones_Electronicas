```
import numpy as np

T=1000e-6 
R=100
t_desfase=1.8*100e-6 #grafico de como lo halle en Word Guia4
vp_c1=5
vp_c2=2.6
error_v1=vp_c1*0.03+0.05
error_v2=vp_c2*0.03+0.05
error_tdfase=(100e-6)/250+t_desfase*(t_desfase*100/1e6)+0.6e-9
error_T=(100e-6)/250+T*(T*100/1e6)+0.6e-9

u_v1=error_v1/np.sqrt(3)
u_v2=error_v2/np.sqrt(3)
u_tdesf=error_tdfase/np.sqrt(3)
u_T=error_T/np.sqrt(3)

#calculo potencia activa
P=vp_c1*(vp_c2/R)*np.cos(t_desfase*2*np.pi/T)

dP_dC1=(vp_c2/R)*np.cos(t_desfase*2*np.pi/T)
dP_dC2=vp_c1*(1/R)*np.cos(t_desfase*2*np.pi/T)
# dP_dtdesf=-vp_c1*(1/R)*np.sin(t_desfase*2*np.pi/T)*2*np.pi/T
# dP_dT=-vp_c1*(1/R)*np.sin(t_desfase*2*np.pi/T)*(-2*np.pi/(T**2))
dP_dtdesf = -(np.pi*vp_c1*vp_c2*np.sin(np.pi*t_desfase/T))/(R*(T))
dP_dT = (np.pi*vp_c1*vp_c2*(t_desfase)*np.sin(np.pi*t_desfase/T))/(R*(T**2))

uP=np.sqrt((u_v1*dP_dC1)**2+(u_v2*dP_dC2)**2+(u_tdesf*dP_dtdesf)**2+(u_T*dP_dT)**2)
k=2
print("Valor potencia activa: %.3f +- %.3f"%(P,uP*k))

#potencia aparente
S=vp_c1*(vp_c2/R)
dS_dC1=vp_c2/R
dS_dC2=vp_c1/R
uS=np.sqrt((u_v1*dS_dC1)**2+(u_v2*dS_dC2)**2)
print("\nValor potencia aparente = %.3f +- %.3f"%(S,k*uS))

#calculo modulo de la impedancia
Z=vp_c1/(vp_c2/R)

Xc=Z*np.sin(t_desfase*2*np.pi/T)
C=T/(Xc*2*np.pi)
print("\nValor del capacitor:",C)

```
