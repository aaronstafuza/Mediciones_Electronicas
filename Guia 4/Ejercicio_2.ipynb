```
import numpy as np
T=1e-3
Vg=5
Vg_RMS=Vg/np.sqrt(2)
Ig_RMS=27.66e-3
Pm = 95e-3 #valor medio de potencia

tita=np.rad2deg(np.arccos(Pm/(Vg_RMS*Ig_RMS)))
print("Tita = %.2fยบ"%tita)

#Calculo de R:
R=(Vg_RMS/Ig_RMS)*np.cos(tita)
print("R = %.2f"%R)

#Calculo de Xc:
Xc=(Vg_RMS/Ig_RMS)*np.sin(tita)
print("Xc = %.2f"%Xc)

#Calculo de C:
C=T/(Xc*2*np.pi)
print("C = ",C)
```
