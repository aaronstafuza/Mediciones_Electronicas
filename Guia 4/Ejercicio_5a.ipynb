```
import numpy as np
V=1.35 #RMS
I=979e-6 #RMS
T=10*200e-6 

Pm=(((2.76e-3)-(380e-6))/2)
print("pot media: ",Pm)

k=2

error_T=((200e-6)/8000+(T*(T*50/1e6))+0.6e-9)/np.sqrt(3)

#incertidumbres:
uc_v=V*0.004
uc_i=I*0.004
uc_Pm=(0.03*Pm)/np.sqrt(3)

#calculo de angulo de desfase
tita=np.rad2deg(np.arccos(Pm/(V*I)))
dTita_dPm=-((np.sqrt(V**2)*np.sqrt(I**2))/(V*I*(np.sqrt((V**2)*(I**2)-(Pm**2)))))
dTita_dV=(Pm*(np.sqrt(V**2)*(np.sqrt(I**2))))/((V**2)*I*(np.sqrt((V**2)*(I**2)-(Pm**2))))
dTita_dI=(Pm*(np.sqrt(V**2)*np.sqrt(I**2)))/(V*(I**2)*(np.sqrt((V**2)*(I**2)-(Pm**2))))
uTita=np.sqrt((dTita_dPm*uc_Pm)**2+(dTita_dV*uc_v)**2+(dTita_dI*uc_i)**2)

print("Angulo desfase = %.2fº +- %.2fº"%(tita,uTita*k))

#Resistencia
R=(V/I)*np.cos(np.deg2rad(tita))
dR_dV=np.cos(tita)/I
dR_dI=-((V*np.cos(tita))/(I**2))
dR_dTita=-((V*np.sin(tita))/(I))
uR=np.sqrt((dR_dV*uc_v)**2+(dR_dI*uc_i)**2+(dR_dTita*uTita)**2)
print("R = %.2f Ω +- %.2f Ω"%(R,uR*2))

#Reactancia capacitiva
Xc=(V/I)*np.sin(np.deg2rad(tita))
dXc_dV=np.sin(tita)/I
dXc_dI=-(V*np.sin(tita))/(I**2)
dXc_dTita=(V*np.cos(tita))/I
uXc=np.sqrt((dXc_dV*uc_v)**2+(dXc_dI*uc_i)**2+(dXc_dTita*uTita)**2)
# print("Xc = %.2f +- %.2f"%(Xc,uXc*2))

#Capacitancia
C=T/(Xc*2*np.pi)
dC_dT=1/(2*np.pi*Xc)
dC_dX=-T/(2*np.pi*(Xc**2))
uC=np.sqrt((error_T*dC_dT)**2+(uXc*dC_dX)**2)

print("C = %.3f µF +- %.3f µF"%(C*1e9,uC*1e9*k))

```
