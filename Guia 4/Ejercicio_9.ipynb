```
import numpy as np

#Ejercicio 9 de guia osciloscopio

PP=10 #punta de prueba
FDV = 50e-3 #Factor de escala del vertical
divs = 4  #cantidad de divisiones contadas en pantalla
CV = FDV*divs #lectura del canal vertical
error_lectura=2*FDV*0.2 #tomamos 2 veces la subdivision de la grilla por ser Vpp 
errorPP=0.02
errorCV=0.03
k=2

T=4*2e-3
f=1/T
u_r_lect=(20/4)/np.sqrt(3)
u_r_fabric=((0.8/100)+(20/4))/np.sqrt(3)
u_r_f=np.sqrt((u_r_fabric**2)+(u_r_lect**2))
print("f = %.1f Hz +- %.3f Hz"%(f,u_r_f*2))

#Vpp = FDV*DIV*PP
#Vpp = CV*PP
Vpp = CV*PP

#errorV=error_CV + error_lectura 
```

```
#incertidumbre absoluta
u_PP = errorPP*Vpp/np.sqrt(3)
u_V = (errorCV*CV + error_lectura)/np.sqrt(3)

dVpp_dV =  PP
dVpp_dPP =  CV

u_vpp = np.sqrt((u_PP*dVpp_dPP)**2+(u_V*dVpp_dV)**2)

print("Valor pico a pico: %0.3f V +- %0.3f V"%(Vpp,u_vpp*2))
```

```
#incertidumbre relativa
u_r_V = (errorCV+error_lectura/CV)/np.sqrt(3)
u_r_PP = errorPP/np.sqrt(3)

u_r_vpp = np.sqrt(u_r_PP**2 + u_r_V**2)

print("Valor pico a pico: %0.3f V +- %0.3f V"%(Vpp,u_r_vpp*Vpp*2))
```

```
#calculo Vp+
Vp_pos=3*FDV*PP
error_lecturaVp=1*FDV*0.2

u_r_V1=(errorCV+error_lectura/CV)/np.sqrt(3)

u_r_vp=np.sqrt((u_r_PP)**2+(u_r_V1)**2) #incertidumbre relativa del valor pico 
print("Valor pico positivo: %.3f V +- %.3f V"%(Vp_pos,u_r_vp*k))

Vp_neg=1*FDV*PP
print("Valor pico negativo: %.3f V +- %.3f V"%(Vp_neg,u_r_vp*k))
```

```
#calculo VRMS
Vrms=np.sqrt(Vp_neg**2+(Vp_pos/np.sqrt(2))**2)

dVrms_dVpNeg=(np.sqrt(2)*Vp_neg)/(np.sqrt((2*Vp_neg)**2+(Vp_pos)**2))
dVrms_dVpPos=(Vp_pos)/(np.sqrt(2)*np.sqrt((2*Vp_neg)**2+(Vp_pos)**2))

uVrms=np.sqrt((u_r_vp*dVrms_dVpNeg)**2+(u_r_vp*dVrms_dVpPos)**2)
print("Valor RMS: %.3f +- %.3f"%(Vrms,uVrms*k))
```
