```
import numpy as np
#para la funcion triangular:
vp_positivo = 4 *(200/1000)
print("Vp+ (triangular): ",vp_positivo)
vp_negativo = 2*(200/1000)
print("Vp- (triangular): ",vp_negativo)

#para la funcion rectangular:
vp2_positivo = 3*(200/1000)
print("\nVp+ (rectangular): ",vp2_positivo)
vp2_negativo = (200/1000)
print("Vp- (rectangular): ",vp2_negativo)

#triangular:
vp1_centrado=3*(200/1000)
Vme1=(vp1_centrado/2)*(np.pi/(2*np.sqrt(2)))
print("\nVme (triangular): ",Vme1)

Vrms1=np.sqrt((vp1_centrado**2)*(0.00004/4))
print("Vrms (triangular): ",Vrms1)

V1=vp1_centrado/np.sqrt(2)
print("Valor indicado por un instrumento que responde al valor pico+ (triangular): ",V1)

#rectangular:
vp2_centrado=2*(200/1000)
Vme2=(vp2_centrado)*(np.pi/(2*np.sqrt(2)))
print("\nVme (rectangular): ",Vme2)

Vrms2=np.sqrt((vp2_centrado**2)*(0.00004))
print("Vrms (rectangular): ",Vrms2)

V2=vp2_centrado/np.sqrt(2)
print("Valor indicado por un instrumento que responde al valor pico+ (rectangular): ",V2)
```
