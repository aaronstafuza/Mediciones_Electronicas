```
import numpy as np
v1=0.57 #valor medido voltimetro 1
v2=1.156
error_1=1/100
error_2=0.5+2/1156*100
k=1/np.sqrt(2) #factor de voltimetro 1

e1=v1*error_1/100
e2=v2*error_2/100

#calculo incertidumbre tipo B
v1j=e1*v1/100
print("Incertidumbre tipo B de V1: ",v1j)
v2j=e2*v2/100
print("Incertidumbre tipo B de V2: ",v2j)

#valor pico
vp=v1/k

#duty
d=(v2**2)/(vp**2+v2**2)
dD_dV2=(2*v2*vp**2)/((vp**2+v2**2)**2)
dD_dVp=-(2*vp*v2**2)/((vp**2+v2**2)**2)
uD=np.sqrt((v1j*dD_dVp)**2+(v2j*dD_dV2)**2)
print("Incertidumbre de duty: ",uD)
k=2
print("Duty con factor de cobertura de 95%%: %f ± %f"%(d,uD*k))

```
