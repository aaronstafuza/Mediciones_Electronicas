```
import numpy as np
v1=2.69   #voltimetro valor maximo
v2=2.026  #voltimetro valor media modulo
v3=-0.807 #voltimetro valor medio (valor dc)
error_1=3/100
error_2=0.5+2/2026*100
error_3=0.25+2/807*100

#errores
e1=error_1*v1/100
e2=error_2*v2/100
e3=error_3*v3/100

#incertidumbre tipo B
v1j=e1/np.sqrt(3)
print("Incertidumbre tipo B V1: ",v1j)
v2j=e2/np.sqrt(3)
print("Incertidumbre tipo B V2: ",v2j)
v3j=e3/np.sqrt(3)
print("Incertidumbre tipo B V3: ",v3j)

#calculo el duty
d=v2/(v1*np.pi)
#calculo incertidumbre del duty
dD_dV1=-v2/(np.pi*v1**2)
dD_dV2=1/(v1*np.pi)
uD=np.sqrt((v1j*dD_dV1)**2+(v2j*dD_dV2)**2)
k=2
print("\nIncertidumbre del ciclo de actividad: ",uD)
print("Duty con factor de cobertura 95%%: %f ± %f"%(d,uD*k))

#calculo los valores pico
vpp=v1*np.sqrt(2)/(1-d)
dVpp_dV1=np.sqrt(2)/(1-d)
dVpp_dD=v1*np.sqrt(2)/((1-d)**2)
uVpp=np.sqrt((v1j*dVpp_dV1)**2+(uD*dVpp_dD)**2)
print("\nIncertidumbre del valor pico a pico: ",uVpp)
print("Vpp con factor de cobertura 95%%: %f ± %f"%(vpp,uVpp*k))


#calculo valor rms en alterna
Va=vpp*np.sqrt((1-d)*d)
dVa_dVpp=np.sqrt((1-d)*d)
dVa_dD=vpp*(1-2*d)/(2*np.sqrt(1-d)*d)
uVa=np.sqrt((uVpp*dVa_dVpp)**2+(uD*dVa_dD)**2)
print("\nIncertidumbre de valor RMS en alterna: ",uVa)
print("Valor RMS con factor de cobertura 95%%: %f ± %f"%(Va,uVa*k))


#calculo valor rms en alterna mas continua
v_rms=np.sqrt(Va**2+v3**2)
dVrms_dVa=2*Va/(2*np.sqrt(Va**2+v3**2))
dVrms_dV3=2*v3/(2*np.sqrt(Va**2+v3**2))
uVrms=np.sqrt((uVa*dVrms_dVa)**2+(v3j*dVrms_dV3)**2)
print("\nIncertidumbre del valor RMS en AC+DC: ",uVrms)
print("Vrms con factor de cobertura de 95%%: %f ± %f"%(v_rms,uVrms))

```
